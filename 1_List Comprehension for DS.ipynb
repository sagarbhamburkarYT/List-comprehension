### List comprehension

Syntax:
> list=[expression for variable in sequence]
> - *expression* is nothing but business logic
> - *variable* can be anything
> - *sequence* can be list, tuple, set, dictionary and range.

<br>
How it works:<br>
1. If expression is satisfied <br>
2. It will collect a data from sequence<br>
3. And execute a for loop<br>
4. And store result into list.<br>


```python
words = ["alpha","omega","up","down","over","under","purple","red","blue","green"]
result = [len(i) for i in words]
result
```

```python
result = len([i for i in words])
result
```

```python
result = sum([len(i) for i in words])
result
```

```python
sentence = "the quick brown fox jumps over the lazy dog"
result = [len(i) for i in sentence.split()] #split()
result
```
